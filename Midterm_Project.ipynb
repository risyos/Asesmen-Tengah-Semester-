# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, f1_score, recall_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

# EDA

```python
df_train = pd.read_csv('emnist-letters-train.csv')
df_train
```

```python
df_train = pd.read_csv('emnist-letters-train.csv',header=None)
df_train
```

```python
df_test = pd.read_csv('emnist-letters-test.csv',header=None)
df_test
```

# Feature Engineering

```python
#fe
print(df_train.shape)
print(df_test.shape)
```

```python
df_merge = pd.concat([df_train, df_test])

samples_per_class = 100

balanced_df = pd.concat([
    df_merge[df_merge[0] == class_id].sample(samples_per_class, random_state=42)
    for class_id in range(1, 27)
])
print(balanced_df)
```

```python
# Mendefinisikan labels_char terlebih dahulu
# Ini untuk EMNIST Letters dataset dimana label 1=A, 2=B, dst
labels_char = {1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'F', 7:'G', 8:'H', 9:'I', 10:'J',
               11:'K', 12:'L', 13:'M', 14:'N', 15:'O', 16:'P', 17:'Q', 18:'R', 19:'S',
               20:'T', 21:'U', 22:'V', 23:'W', 24:'X', 25:'Y', 26:'Z'}

# Menampilkan 5 gambar
plt.figure(figsize=(10, 2))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(balanced_df.iloc[i, 1:].values.reshape(28, 28).T, cmap='gray')

    # Ambil label dari kolom pertama (indeks 0) untuk setiap baris
    label_num = int(balanced_df.iloc[i, 0])

    # Konversi label numerik ke huruf
    if label_num in labels_char:
        label_char = labels_char[label_num]
        plt.title(f'Label: {label_char}')
    else:
        plt.title(f'Label: {label_num}')

    plt.axis('off')

plt.suptitle("Contoh 5 Gambar EMNIST")
plt.tight_layout()
plt.show()
```

```python
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
balanced_df
```

```python
# Ekstrak fitur dan label
x = balanced_df.iloc[:, 1:].values  #fitur (gambar)
y = balanced_df.iloc[:, 0].values   #label (kelas huruf A-Z)

# shuffle dan bagi 80% train, 20% test
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=42)
```

# HOG

```python
# Konfigurasi parameter HOG
hog_params = {
    'orientations': 6,
    'pixels_per_cell': (4, 4),
    'cells_per_block': (4, 4),
    'block_norm': 'L2'
}

# Fungsi untuk ekstraksi fitur HOG
def extract_hog_features(images):
    features = []
    for row in images:
        image = row.reshape(28, 28).T  # Transpose agar orientasi benar
        hog_feature = hog(image, **hog_params)
        features.append(hog_feature)
    return np.array(features)

# Ekstrak fitur HOG dari data training dan testing
# Pastikan x_train dan x_test sudah didefinisikan sebelumnya
x_train_hog = extract_hog_features(x_train)
x_test_hog = extract_hog_features(x_test)

# Tampilkan ukuran fitur HOG
print(f"HOG feature size (train): {x_train_hog.shape}")
print(f"HOG feature size (test): {x_test_hog.shape}")
```

```python
clf = SVC(
    kernel ='poly',
    C = 10.0,
    gamma = 1
)
```

# Training

```python
clf.fit(x_train_hog, y_train)
```

```python
y_pred = clf.predict(x_test_hog)
```

# Evaluation

```python
print("\nclassification Report:\n", classification_report(y_test, y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
print("F1 Score (macro):", f1_score(y_test, y_pred, average='macro'))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision (macro): {:.2f}".format(precision_score(y_test, y_pred, average='macro')))
print("Recall (macro): {:.2f}".format(recall_score(y_test, y_pred, average='macro')))
print("F1 Score (macro): {:.2f}".format(f1_score(y_test, y_pred, average='macro')))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
y_pred = clf.predict(x_test_hog)

print("=== Evaluation on test set ===")
# Ganti y_test_pred dengan y_pred
print("\nclassification_report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision (macro):", recall_score(y_test, y_pred, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
print("F1 Score (macro):", f1_score(y_test, y_pred, average='macro'))

# Ganti y_test_pred dengan y_pred
cm_test = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_test, annot=False, cmap='Greens')
plt.title('Confusion Matrix - Test Set')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

# Hypertuning

```python
X_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train_hog = extract_hog_features(X_train)
x_test_hog = extract_hog_features(x_test)

param_grid = {
    'C': [1, 2, 3],
    'gamma' : [0.01, 0.02, 0.03],
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid']
}
```

```python
svc = SVC()
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, scoring='accuracy', cv=10, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(X_train_hog, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross:.Validation Accuracy:", grid_search.best_score_)
```
